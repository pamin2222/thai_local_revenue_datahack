### Read in all files as list

```python
revenue64_file_path = os.path.join(data_dir, "รายได้ อปท. ปี 61-65", "รายได้ อปท. ปี 61-65 - 2564.csv")
revenue65_file_path = os.path.join(data_dir, "รายได้ อปท. ปี 61-65", "รายได้ อปท. ปี 61-65 - 2565.csv")
revenue63_file_path = os.path.join(data_dir, "รายได้ อปท. ปี 61-65", "รายได้ อปท. ปี 61-65 - 2563.csv")
revenue62_file_path = os.path.join(data_dir, "รายได้ อปท. ปี 61-65", "รายได้ อปท. ปี 61-65 - 2562.csv")
revenue61_file_path = os.path.join(data_dir, "รายได้ อปท. ปี 61-65", "รายได้ อปท. ปี 61-65 - 2561.csv")
data_filepath_list = [revenue61_file_path, revenue62_file_path, revenue63_file_path, revenue64_file_path, revenue65_file_path]
```

```python
def read_all_df(data_filepath_list):
    df_list = []
    for data_filepath in data_filepath_list:
        # Extract the year from the file name
        data_year = int(data_filepath.split("-")[-1].split(".")[0].strip())
        #skip last row (total row)
        if '2563' in data_filepath: #2563 has no total row
            data_df_temp = pd.read_csv(data_filepath, engine='python', encoding='utf-8')
        else:
            data_df_temp = pd.read_csv(data_filepath, skipfooter=1, engine='python', encoding='utf-8')
        # Remove whitespace from column names
        data_df_temp.columns = data_df_temp.columns.str.strip()
        data_df_temp['data_year'] = data_year
        data_df_temp['row_id'] = data_df_temp['ลำดับ'].astype(str).str.strip() + "_" + str(data_year)
        data_df_temp = data_df_temp.set_index(['row_id'])
        df_list.append(data_df_temp.copy())
    return df_list
df_list = read_all_df(data_filepath_list)
print(len(df_list))
```
